```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
##distribution of label for amazon turk 
df = pd.read_csv("/content/drive/MyDrive/fyp/fyp2/amazon turks concat/full_amazon.csv")
df_test = "/content/drive/MyDrive/fyp/fyp2/final models and datasets/dataset/test_dataset.csv"
df_train ="/content/drive/MyDrive/fyp/fyp2/final models and datasets/dataset/train_dataset.csv"
df_valid = "/content/drive/MyDrive/fyp/fyp2/final models and datasets/dataset/valid_dataset.csv"
save_folder = "/content"
  
def label_dis_graph(df,name,save_folder=None,):
  df_pos = df[df['label'] == 'Positive']
  df_neu = df[df['label'] == 'Neutral']
  df_neg = df[df['label'] == 'Negative']

  pos = df_pos.shape[0]
  neu = df_neu.shape[0]
  neg = df_neg.shape[0]
  print("values: pos ",pos," neg: " , neg , "neu:" , neu)

  #distri_class = pd.DataFrame({'class':['Positive', 'Neutral', 'Negative'], 'size':[pos, neu,neg]})

  #ax = distri_class.plot.bar(x='class', y='size', rot=0)

  fig, axs = plt.subplots(1,2)
  fig.suptitle(name)
  axs[0].set_title('Bar Chart')
  axs[1].set_title('Pie Chart')
  #plt.title('Bar Chart showing distribution of labels in labeled amazon dataset')

  axs[0].bar( ['Positive', 'Neutral', 'Negative'],[pos,neu,neg])
  axs[1].pie([pos,neu,neg], labels = ['Positive', 'Neutral', 'Negative'])
  if save_folder != None:
    plt.savefig(save_folder + "/" + name + ".png", dpi=fig.dpi)
    print('saved')
  plt.show()
  return plt


 

label_dis_graph(df,'Graphs showing distribution of labels in labeled amazon dataset')

```
