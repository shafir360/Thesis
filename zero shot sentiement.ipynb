```python
import pandas as pd

dataset = pd.read_csv("/content/drive/MyDrive/fyp/fyp2/labelled data concat/full_labelled_data.csv")

test_set_df  = dataset
display(test_set_df)
```

```python

from transformers import pipeline

classifier = pipeline("zero-shot-classification")
```

```python
the_labels = ["Positive", "Negative","Neutral","spam"]
a_review = "The movie was so boring, I actually fell sleep in the first 20 minutes."

res = classifier(a_review, the_labels)


print( res )
```

```python
from tqdm import tqdm
from sklearn.metrics import accuracy_score

results = []
targets = []

for idx, item in tqdm(test_set_df.iterrows()):
  
  res = classifier(item['text'], the_labels)

  results.append( res['labels'][0] )
  targets.append( item['label'] )
  
accuracy = accuracy_score(results, targets)
```

```python
print(accuracy)
```
