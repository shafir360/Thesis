```python
##update vader dictionary
from nltk.tokenize import word_tokenize, RegexpTokenizer

import pandas as pd

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


analyzer = SentimentIntensityAnalyzer()
sentence = "another bear market is about to come"

new_words = {
    'bull': 2,
    'bear': -2,
    'bullish': 2,
    'bearish': -2,
    'dump': -1.5,
    'crash':-1.7,
    'crashing':-1.5

}

analyzer.lexicon.update(new_words)
s = analyzer.polarity_scores(sentence)


def pseudo_vader_based_labelling(tweet):
  #print(tweet)
  scores = analyzer.polarity_scores(tweet)
  #print(scores)
  score = scores['compound']
  
  if score > 0:
    return "Positive"
  elif score < 0:
    return "Negative"
  elif score == 0:
    return "Neutral"
  else:
    print("Error at pseudo_vader_based_labelling")
    return 

```

```python
unused_df = new_df.copy()

unused_df = unused_df.dropna(subset = ['tweet'])
unused_df['vader_pseudo_label'] = unused_df.tweet.apply(pseudo_vader_based_labelling)
unused_df.to_csv("/content/drive/MyDrive/fyp/fyp2/unused_Bitcoin_tweets_cleaned_2.csv")
```
