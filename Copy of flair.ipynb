```python
#data_columns = ["target", "ids", "date", "flag", "user", "text"]
#data_encoding = "ISO-8859-1"
#data_file = pd.read_csv('/content/drive/MyDrive/fyp/fyp2/training.1600000.processed.noemoticon.csv',encoding = data_encoding, names=data_columns)
data_file_o = pd.read_csv('/content/drive/MyDrive/fyp/fyp2/Bitcoin_tweets_cleaned_no_duplicates.csv')
data_file_o.replace("", np.nan, inplace=True) # replace empty string with np.nan
data_file_o.dropna(subset = ["tweet"], inplace=True) # drop nan values
import pandas as pd
import flair
from flair.models import TextClassifier
from flair.data import Sentence
```

```python

data_file = data_file_o.copy()

sia = TextClassifier.load('en-sentiment')


def sentiment_Flair(x):
    if x == '' or x == "" or x == np.nan:
      print("done")
      return np.nan,np.nan
    else:
      try:
        #print(x)
        sample = Sentence(x)
        sia.predict(sample)
        #print(sample)
        #print(i)
        #print(".")
        #print(sample.labels[0])
        return sample.labels[0].value ,sample.labels[0].score
      except:
        return np.nan,np.nan


#data_file['sentiment_flair'] = data_file['tweet'].apply(lambda x: sentiment_Flair(x))
#data_file['sentiment_flair_conf'] = data_file['tweet'].apply(lambda x: sentiment_conf(x))

df_label = []
df_conf = []

for tweet in data_file['tweet']:
  label,conf = sentiment_Flair(tweet)
  df_label.append(label)
  df_conf.append(conf)



data_file['label'] = df_label
data_file['conf'] = df_conf
display(data_file)
```
