```python
df = pd.read_csv('/content/drive/MyDrive/fyp/fyp2/Bitcoin_tweets_cleaned.csv')
df.replace("", np.nan, inplace=True) # replace empty string with np.nan
df.dropna(subset = ["tweet"], inplace=True) # drop nan values
```

```python
def huggingface_model_out(directory,tweets):

  tokenizer = AutoTokenizer.from_pretrained(directory)
  model = AutoModelForSequenceClassification.from_pretrained(directory)
  classifier = pipeline("sentiment-analysis", model = model, tokenizer = tokenizer)
  df = pd.DataFrame(columns=['label','conf'])
  for tweet in tweets:
    res = classifier(tweet)[0]
    #print(res)
    #print(type(res))
    label = res['label']
    #print(label)
    score = res['score']
    #print(score)
    new_row = {'label':label, 'conf':score}
    #print(new_row)
    df = df.append(new_row, ignore_index=True)
    
  return df

model_name = "svalabs/twitter-xlm-roberta-bitcoin-sentiment"


#tokenizer = AutoTokenizer.from_pretrained(model_name)
#model = AutoModelForSequenceClassification.from_pretrained(model_name)
#classifier = pipeline("sentiment-analysis", model = model, tokenizer = tokenizer)
#res = classifier(df.tweet.values.tolist())

result = huggingface_model_out(model_name,df['tweet'])
#result = huggingface_model_out(model_name,df['tweet'])



#display(result)
```
