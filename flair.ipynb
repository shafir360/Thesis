```python

data_file = data_file_o.copy()

sia = TextClassifier.load('en-sentiment')


def sentiment_Flair(x):
    if x == '' or x == "" or x == np.nan:
      print("done")
      return np.nan
    else:
      try:
        print(x)
        sample = Sentence(x)
        sia.predict(sample)
        #print(sample)
        print(i)
        print(".")
        #print(sample.labels[0])
        return sample.labels[0].value
      except:
        return np.nan

def sentiment_conf(x):

    if x == '' or x == "" or x == np.nan:
      print("done")
      return np.nan
    else:
      try:
        print(x)
        sample = Sentence(x)
        sia.predict(sample)
        #print(sample)
        print(i)
        print(".")
        #print(sample.labels[0])
        return sample.labels[0].score
      except:
        return np.nan

i= 0 
for tweet in data_file['tweet']:
  print(i)
  sentiment_Flair(tweet)
  i = i + 1

bob  =sentiment_Flair("")
print(bob)
print("printed bob")
data_file['sentiment_flair'] = data_file['tweet'].apply(lambda x: sentiment_Flair(x))
data_file['sentiment_flair_conf'] = data_file['tweet'].apply(lambda x: sentiment_conf(x))
# write file
#csv_data = data_file.to_csv('E:/MS Thesis/Dataset MS Thesis/ALL DATA SQL/All Data For all Wave/flair_sentiment.csv')
display(data_file)
```
