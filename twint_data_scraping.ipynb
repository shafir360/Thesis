```python
import twint
import nest_asyncio
nest_asyncio.apply()


output_folder = '/content/drive/MyDrive/fyp/fyp2/twint3'
name = 'twint1'

out = output_folder + "/" + name + ".csv"

def main():
  # Configure
  c = twint.Config()
  c.Resume = output_folder + '/resume_file.txt'
  c.Search = "Bitcoin OR btc AND -nft AND -giveaway AND -guiding AND -guided AND -#nft AND -#btc AND -#bitcoin"
  c.Lang = 'en'
  #c.Store_csv = True
  c.Links = 'exclude' 
  c.Filter_retweets = True        
  #c.Limit = 20   
  c.Count = True     
  c.Filter_retweets = True
  c.Store_csv = True
  c.Hide_output = True
  c.Custom["tweet"] = ["user_id","id","username", "created_at","date","time","timezone","tweet","language","retweets_count","hashtags","retweet"]
  c.User_full = True
  c.Count = True
  c.Output =  out 
  # Run   
  twint.run.Search(c)                   

while True:
    try:
        main()
        break # stop the loop if the function completes sucessfully
    except Exception as e:
        print("Function errored out!", e)
        print("Retrying ... ")

df = pd.read_csv(out)
display(df)
```
